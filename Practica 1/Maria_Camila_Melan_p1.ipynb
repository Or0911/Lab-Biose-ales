```python
# j) Construir la funci贸n y[n] = sin( * 0.12 * n)
n_values = np.arange(0, 101)
y_n = np.sin(np.pi * 0.12 * n_values)

plt.plot(y_n)
plt.title('Gr谩fica de y[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()
```

```python
# k) Construir otra senoidal y2[n] = cos(2 * 0.03 * n)
y2_n = np.cos(2 * np.pi * 0.03 * n_values)
plt.plot(y2_n)
plt.title('Gr谩fica de y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()
```

```python
# l) Generar se帽ales s[n] = y[n] + y2[n] y t[n] = y[n] * y2[n]
s_n = y_n + y2_n
t_n = y_n * y2_n

# grafica s_n
plt.subplot(2,1,1)
plt.plot(s_n)
plt.title('Gr谩fica de s[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()

# grafica s_n
plt.subplot(2,1,2)
plt.plot(t_n)
plt.title('Gr谩fica de t[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()
```

```python
# m) Graficar y[n] y y2[n]
plt.figure()
plt.plot(n_values, y_n, label='y[n] = sin( * 0.12 * n)')
plt.plot(n_values, y2_n, label='y2[n] = cos(2 * 0.03 * n)')
plt.legend()
plt.title('Gr谩fica de y[n] y y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()
```

```python
# n) Graficar s[n] y t[n]
plt.figure()
plt.plot(n_values, s_n, label='s[n] = y[n] + y2[n]')
plt.plot(n_values, t_n, label='t[n] = y[n] * y2[n]')
plt.legend()
plt.title('Gr谩fica de s[n] y t[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.show()
```

```python
import pandas as pd
import numpy as np

def obtener_estadisticas_notas(notas_alumnos):
    # Convertir el diccionario a una serie de Pandas
    series_notas = pd.Series(notas_alumnos)

    # Calcular estad铆sticas
    nota_minima = series_notas.min()
    nota_maxima = series_notas.max()
    media = series_notas.mean()
    desviacion_tipica = series_notas.std()

    # Crear una nueva serie con las estad铆sticas
    estadisticas = pd.Series({
        'Nota M铆nima': nota_minima,
        'Nota M谩xima': nota_maxima,
        'Media': media,
        'Desviaci贸n T铆pica': desviacion_tipica
    })

    return estadisticas

    # Ejemplo de uso
notas_alumnos = {'Alumno1': 8.5, 'Alumno2': 7.2, 'Alumno3': 9.0, 'Alumno4': 6.8, 'Alumno5': 8.9}
resultado = obtener_estadisticas_notas(notas_alumnos)

print(resultado)
```
