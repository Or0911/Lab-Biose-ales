Construya la funci√≥n ùë¶[ùëõ]=sin‚Å°(ùúã‚àó0.12ùëõ) en el intervalo 0 <= n <= 100.

```python
n=np.arange(0,101)
y1=np.sin(np.pi*0.12*n)
y2=np.cos(2*np.pi*0.03*n)
s=y1+y2
mult=y1*y2
```

Graficar en la misma figura las se√±ales y[n] y y2[n]

```python
ax=plt.subplot()
ax.plot(y1)
ax.plot(y2)
plt.xlabel("n")
plt.ylabel("y")
plt.legend
plt.show()
```
