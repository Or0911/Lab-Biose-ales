**1. Contexto:**

**Consultar historia, aplicaciones y principales industrias de prótesis mioeléctricas**

Las prótesis mioeléctricas son dispositivos protésicos diseñados para restaurar la función  de una extremidad amputada utilizando señales eléctricas generadas por los músculos residuales del paciente. 

**_Historia y evolución de las prótesis mioeléctricas:_**

La idea de utilizar señales eléctricas musculares para controlar dispositivos protésicos se remonta a la década de 1940. En 1948, Reinhold Reiter, un médico alemán, desarrolló el primer sistema de control mioeléctrico para una prótesis de mano [1]. Sin embargo, no fue hasta la década de 1960 que las prótesis mioeléctricas se hicieron más viables gracias a los avances en la electrónica y la miniaturización de los componentes [2]. 
Durante las décadas de 1970 y 1980, se realizaron importantes investigaciones y desarrollos en el campo de las prótesis mioeléctricas, con la introducción de sistemas de control más sofisticados y la mejora de los materiales y diseños [3].

**_Aplicaciones de las prótesis mioeléctricas:_**

Las prótesis mioeléctricas se utilizan principalmente en personas con amputaciones de miembros superiores (brazos y manos). Permiten un control más natural y una mayor destreza en la realización de tareas cotidianas, como agarrar objetos, manipular herramientas y realizar movimientos finos [4]. 

Algunas de las funcionalidades clave de las prótesis mioeléctricas modernas incluyen:

Control proporcional: La fuerza de agarre se puede ajustar proporcionalmente a la intensidad de la señal mioeléctrica.

Múltiples grados de libertad: Las prótesis pueden controlar varios movimientos, como la flexión de la muñeca, la rotación del antebrazo y el agarre de los dedos.

Retroalimentación sensorial: Algunas prótesis incorporan sensores que brindan retroalimentación táctil o de fuerza al usuario.

Patrones de agarre: Las prótesis pueden realizar patrones de agarre predefinidos, como agarre de pinza o agarre de gancho [4].

Además de su uso en prótesis de miembros superiores, las señales mioeléctricas también se han utilizado para controlar prótesis de miembros inferiores, sillas de ruedas eléctricas y otros dispositivos de asistencia [5]. 

**_Industrias principales de prótesis mioeléctricas:_**

La industria de las prótesis mioeléctricas ha experimentado un crecimiento significativo en las últimas décadas. Algunas de las principales empresas y organizaciones involucradas en el desarrollo y fabricación de estas prótesis son:

Empresas de dispositivos médicos:

Liberating Technologies (Estados Unidos): Compañía enfocada en el desarrollo de prótesis mioeléctricas y tecnologías de interfaz neural [9]. 

Instituciones académicas y de investigación: 

Universidades y centros de investigación desempeñan un papel crucial en el avance de las tecnologías de prótesis mioeléctricas. Algunos ejemplos notables incluyen el MIT Media Lab, la Universidad de Nuevo México y el Instituto de Biomecánica de Valencia [10]-[11].

**_Estrategias para mejorar la fijación de los sensores EMG:_**

Utilizar sistemas de fijación ajustables y cómodos: Los sensores EMG deben estar bien sujetos a la piel para evitar desplazamientos y minimizar el ruido en la señal. Se pueden utilizar sistemas de fijación ajustables, como brazaletes elásticos o mangas de compresión, que mantengan los sensores en su lugar sin causar molestias [14]. 

Emplear electrodos secos o de gel sólido: Los electrodos secos o de gel sólido pueden proporcionar una mejor adherencia a la piel y reducir el riesgo de desprendimiento en comparación con los electrodos húmedos tradicionales [15]. 

Preparación adecuada de la piel: Antes de colocar los electrodos, es importante preparar la piel mediante la limpieza y el afeitado del área de interés para mejorar la conductividad y la adherencia de los sensores [16].

**_Ubicación óptima del electrodo para capturar el movimiento de flexión-extensión del codo:_**

El bíceps  y el tríceps son los músculos principales involucrados en la flexión y extensión del codo, respectivamente. Varios estudios han investigado la ubicación óptima de los electrodos para capturar estas señales musculares. 

Según Kendell et al. [17], la ubicación recomendada para el bíceps braquial es aproximadamente a 1/3 de la distancia entre el hombro y el codo, en la línea entre la fosa cubital y el acromion. Para el tríceps braquial, los electrodos deben colocarse a aproximadamente 1/2 de la distancia entre el hombro y el codo, en la línea entre el olécranon y el acromion. 


Para la flexión del codo (bíceps ): La posición estándar es colocar el electrodo en la porción más prominente del centro muscular del bíceps , aproximadamente a 1/3 de la distancia entre el hombro (acromion) y el codo (fosa cubital) [18][19]. 

Para la extensión del codo (tríceps braquial): El electrodo debe colocarse sobre la porción más prominente del centro muscular del tríceps, aproximadamente a 1/2 de la distancia entre el hombro (acromion) y el codo (olécranon) [18][19]. 

Ahora bien, para capturar las señales electromiográficas (EMG) de los músculos del antebrazo para movimientos de flexión-extensión de la muñeca, las ubicaciones recomendadas de los electrodos son las siguientes: 

Flexión de la muñeca: Los electrodos deben colocarse sobre el centro muscular del flexor radial del carpo y el palmar largo, ubicados en la región anterior del antebrazo [20][21]. 

Extensión de la muñeca: Los electrodos se deben colocar sobre el centro muscular del extensor radial largo del carpo y el extensor común de los dedos, ubicados en la región posterior del antebrazo [22][23]. 

Ubicaciones específicas recomendadas:

Flexor radial del carpo: Aproximadamente a 1/3 de la distancia entre el codo y la muñeca, en la cara anterior del antebrazo. 

Palmar largo: Aproximadamente a 1/2 de la distancia entre el codo y la muñeca, en la cara anterior del antebrazo. 

Extensor radial largo del carpo: Aproximadamente a 1/4 de la distancia entre el codo y la muñeca, en la cara posterior del antebrazo. 

Extensor común de los dedos: Aproximadamente a 1/3 de la distancia entre el codo y la muñeca, en la cara posterior del antebrazo [24].

```python
# Función para aplicar el operador Teager-Kaiser Energy Operator (TKEO) a una señal
def tkeo(signal):
    tkeo_signal = signal**2 - np.roll(signal, 1) * np.roll(signal, -1)
    return tkeo_signal
```

```python
def apply_tkeo(data):
    tkeo_data = data.copy()
    tkeo_data['EMG'] = tkeo(tkeo_data['EMG'])
    return tkeo_data
```

```python
# Datos aplicando el operador
df_all_tkeo = apply_tkeo(df_all)
df_sin_tkeo = apply_tkeo(df_sin) 
```

```python
# Función para calcular diversas características de una señal EMG
def extract_features1(signal):
    rms = np.sqrt(np.mean(signal**2))
    variance = np.var(signal)
    mean_absolute_value = np.mean(np.abs(signal))
    mean_absolute_value_slope = np.mean(np.abs(np.diff(signal)))
    slope_sign_changes = np.sum(np.diff(np.sign(np.diff(signal))) != 0)
    wave_form_length = np.sum(np.abs(np.diff(signal)))
    zero_crossings = np.sum(np.diff(np.sign(signal)) != 0)/2

    return {
        'RMS': rms,
        'Variance': variance,
        'Mean Absolute Value': mean_absolute_value,
        'Mean Absolute Value Slope': mean_absolute_value_slope,
        'Slope Sign Changes': slope_sign_changes,
        'Wave Form Length': wave_form_length,
        'Zero Crossings': zero_crossings
    }
```

**Independencia:**

Se asumen ambos movimientos como independientes, dado que se tomaron en momentos diferentes, y el uno no depende del otro. 

**Normalidad**

Para evaluar este supuesto se parte de las siguientes hipótesis 
H0 = La distribución de los datos es normal.

H1 = La distribución de los datos no es normal.

Se calcula el valor p;

se define un α = 0.05

Si bien las diferencias observadas a simple vista pueden parecer relevantes, es necesario respaldarlas con pruebas estadísticas que tengan en cuenta la variabilidad inherente de los datos y proporcionen evidencia cuantitativa de la significancia estadística de esas diferencias.
